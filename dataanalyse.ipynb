```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
df = pd.read_csv('/content/Employee Performance Data.csv')
print(df.head())
print(df.shape)
```

```python
print(df.describe)
print(df.tail)
```

```python
print('Missing values/n',df.isnull().sum())
```

```python
df.hist(bins=50, figsize=(20,15))
plt.show
```

```python
corr = df.corr()
plt.figure(figsize = (10,10))
plt.imshow(corr,cmap = 'twilight_shifted', interpolation = 'none')
plt.colorbar()
plt.xticks(range(len(corr)),corr.columns)
plt.yticks(range(len(corr)),corr.columns)
plt.show()
```

```python
plt.figure(figsize=(10,9))
plt.scatter(df['Age'],df['Monthly_Salary_USD'])
plt.xlabel('Age')
plt.ylabel('Monthly_Salary_USD')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

x = df.drop('Monthly_Salary_USD',axis=1)
y = df['Monthly_Salary_USD']
x_train , x_test ,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test,y_pred)
r2 = model.score(x_test,y_test)
print('R-squared:',r2)
print('Mean Squared Error:',mse)
print('/nModel Evaluation')
print('R-squared:',r2)
print('Mean Squared Error:',mse)
```

```python
plt.figure(figsize=(10,9))
plt.scatter(y_test , y_pred,color = 'green')
plt.plot(y_test , y_test ,color = 'red')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
```
